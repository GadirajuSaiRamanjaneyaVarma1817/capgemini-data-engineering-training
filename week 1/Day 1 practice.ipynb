### Read the file from catalog

```python
df = spark.read.csv("/Volumes/workspace/default/de/C1.csv",header=True,inferSchema=True)
```

```python
df.display()
```

### Handle Blank Values

```python
from pyspark.sql.functions import *
df = df.replace(["blank","Blank"],None)
```

```python
df.display()
```

### Fix country columns

```python
df = df.withColumn("Country",upper(col("Country")))
```

```python
df.display()
```

### Fill missing values

```python
df = df.fillna({
    "Category":"Unknown",
    "Sales":0
})
```

```python
df.display()
```

### Remove Duplicates

```python
df = df.dropDuplicates((["Name"]))
df.display()
```
